```python
map_item_to_twu = {}
map_item_to_twu0 = {}
sum_twu = 0
min_util = 40
database = []
database_trans = []
trans_count = 0
bucket_num = 2
scout_bees_bucket = []
iterations = 10


class Pair:
    def __init__(self, it, ut):
        self.item = it
        self.utility =ut
        self.rutil = 0
        
class Item:
    def __init__(self, item):
        self.item = item
        self.TIDS = []


with open("abc_test.txt", "r") as f:
    file = f.readlines()
    
    for this_line in file:
        if this_line == "" or this_line[0]=="#" or this_line[0]=="%" or this_line[0]=="@":
            continue
            
        splt = this_line.split(":")
        items = list(map(int, splt[0].split(" ")))
        trans_util = int(splt[1])
        sum_twu += trans_util
        
        for item in items:
            
            if item in map_item_to_twu:
                map_item_to_twu[item] += trans_util
            else:
                map_item_to_twu[item] = trans_util
                
            if item in map_item_to_twu0:
                map_item_to_twu0[item] += trans_util
            else:
                map_item_to_twu0[item] = trans_util
                
        util_values = list(map(int, splt[2].split(" ")))
        
        revised_trans = []
        pattern = []
        
        remaining_util = 0
        
        for i in range(len(items)):
            if map_item_to_twu.get(items[i]) >= min_util:
                revised_trans.append(Pair(items[i], util_values[i]))
                pattern.append(items[i])
                remaining_util += util_values[i]
            else:
                map_item_to_twu0.pop(items[i])
                
        for pair in revised_trans:
            remaining_util -= pair.utility
            pair.rutil = remaining_util
# for pair in revised_trans:
#     print(pair.item, pair.utility, pair.rutil)
        
        database.append(revised_trans)
        database_trans.append(pattern)
        trans_count += 1
        
# for i in database:
#     for pair in i:
#         print(pair.item, pair.utility, pair.rutil)

twu_pattern = list(map_item_to_twu0)
twu_pattern.sort()
print(twu_pattern)

m = int(len(twu_pattern)/bucket_num)
items = []

for it in twu_pattern:
    items.append(items(it))
    
for i in range(len(database)):
    for j in range(len(items)):
        for k in range(len(database[i])):
            if items[j] == database[i][j].item:
                items[j].TIDS.append(i)
                
for i in range(len(bucket_num)):
    scout_bees_bucket.append(1)
    
    
if len(twu_pattern) > 0:
    intialization(min_utility)
    for i in range(iterations):
        init_bit_no = 32+1
        employedBees(min_util)
        calcRfitness()
        onlookerBees()
        calScoutBees()
        scoutBees(init_bit_no, min_util)
```
